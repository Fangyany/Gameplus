```python
import pandas as pd
import matplotlib.pyplot as plt
# import seaborn as sns
# sns.set()

def plot_loss(loss_name, key):
    loss_data = pd.read_csv(f'training_log/Exp{loss_name}/train_log.csv') 
    loss = loss_data[key]
    plt.plot(loss, label=f'method_{loss_name}')
    plt.legend()
    # return loss

for key in ['val-planningADE', 'val-planningFDE', 'val-planningAHE', 
            'val-planningFHE', 'val-predictionADE', 'val-predictionFDE']:
   
    plt.rcParams['figure.figsize'] = (5, 5)
    for loss_name in ['_dual', '_dual_fusion', '_dual_add']:
        plot_loss(loss_name, key)
        
    loss_data = pd.read_csv(f'/data/fyy/GameFormer-Planner/training_log/Exp2_40/train_log.csv') 
    loss = loss_data[key]
    plt.plot(loss, label='method_EXP1')
    
    # loss_data = pd.read_csv(f'training_log/Exp_pre_nuplan/train_log.csv') 
    # loss = loss_data[key]
    # plt.plot(loss, label='method_EXP0')
        
    plt.legend()
    plt.ylabel(f'{key}')
    plt.xlabel('Iteration Number')

    plt.show()

# plt.savefig('fig/pred_loss.eps', format='eps', dpi=300)
```
